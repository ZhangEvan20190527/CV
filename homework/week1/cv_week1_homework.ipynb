```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("week1_homework.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img_out = cv2.bilateralFilter(src=img, d=0, sigmaColor=100, sigmaSpace=15)
plt.imshow(img_out)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, binary = cv2.threshold(img_gray, 85, 255, cv2.THRESH_BINARY)
plt.imshow(binary)
```

```python
import numpy as np
img_final = np.ones([1000, 707, 3], dtype=np.int)
for x in range(1000):
    for y in range(707):
        if img_gray[x, y] > 85:
            for z in range(3):
                img_final[x, y, z] = int(img_out[x, y, z])
        else:
            for z in range(3):
                img_final[x, y, z] = int(img[x, y, z])
plt.imshow(img_final)
```
